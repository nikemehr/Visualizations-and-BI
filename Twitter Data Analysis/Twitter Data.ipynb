```python
sum_by_country = sum_by_country[sum_by_country['loc_contribution'] > 1000]
sum_by_country
```

```python
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
def get_continent(col):
    try:
        cn_a2_code =  country_name_to_country_alpha2(col)
    except:
        cn_a2_code = 'Unknown' 
    try:
        cn_continent = country_alpha2_to_continent_code(cn_a2_code)
    except:
        cn_continent = 'Unknown' 
    return (cn_a2_code, cn_continent)
```

```python
get_continent('Australia')
```

```python
for item in sum_by_country['Country']:
    sum_by_country['Code'] = get_continent(item)[0]
    sum_by_country['Continent'] = get_continent(item)[1]

sum_by_country
```

```python
sum_by_country['Code'] = sum_by_country['Country'].apply(lambda x: get_continent(x)[0])
sum_by_country['Continent'] = sum_by_country['Country'].apply(lambda x: get_continent(x)[1])
sum_by_country
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="niknik")
def geolocate(country):
        loc = geolocator.geocode(country)
        return (loc.latitude, loc.longitude)
```

```python
sum_by_country['Latitude'] = sum_by_country['Country'].apply(lambda x: geolocate(x)[0])
sum_by_country['Longitude'] = sum_by_country['Country'].apply(lambda x: geolocate(x)[1])
sum_by_country
```

```python
import folium
from folium.plugins import MarkerCluster
#empty map
world_map= folium.Map(tiles="cartodbpositron")
marker_cluster = MarkerCluster().add_to(world_map)
#for each coordinate, create circlemarker of user percent
for i in range(len(sum_by_country)):
        lat = sum_by_country.iloc[i]['Latitude']
        long = sum_by_country.iloc[i]['Longitude']
        radius=5
        popup_text = """Country : {}<br>
                    %of Users : {}<br>"""
        popup_text = popup_text.format(sum_by_country.iloc[i]['Country'],
                                   sum_by_country.iloc[i]['loc_contribution']
                                   )
        folium.CircleMarker(location = [lat, long], radius=radius, popup= popup_text, fill =True).add_to(marker_cluster)
```
